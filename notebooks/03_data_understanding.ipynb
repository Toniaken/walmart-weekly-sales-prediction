- **0.0% missing data**

This means the dataset is complete and no missing-value treatment is currently required.

Therefore, techniques such as dropping rows, filling values with the mean, median, or mode, or using an imputer are not needed at this stage.

However, missing-value checks will still remain part of the preprocessing pipeline so that the final model can handle unexpected missing inputs more safely during deployment.

```python
# Inspect the date values

date_preview = pd.to_datetime(
    df["Date"],
    format="%d-%m-%Y",
    errors="coerce",
)

print("Earliest date:", date_preview.min())
print("Latest date:", date_preview.max())
print("Invalid dates:", date_preview.isna().sum())
```

### Date-Range Interpretation

The date preview shows that the dataset covers the period from **5 February 2010** to **26 October 2012**.

A total of **0 invalid dates** were found, which means every value in the `Date` column was successfully interpreted using the expected `day-month-year` format.

This confirms that the date values are consistent and suitable for later conversion to a proper datetime data type.

Because the observations are time-based, the dataset should later be sorted by date and split chronologically for model training and evaluation. This will help prevent future information from leaking into the training data.
