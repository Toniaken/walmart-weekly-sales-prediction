### Resolution

2. Installed all project dependencies from `requirements.txt`.

4. Selected the project's `.venv` interpreter as the notebook kernel.

### Lesson Learned
Always verify that Jupyter is using the project's virtual environment rather than the global Python installation. This prevents package import errors and ensures a reproducible development environment.

```python
# Verify that the file exists

if not raw_data_path.exists():
    raise FileNotFoundError(
        f"The Walmart dataset was not found at: {raw_data_path}"
    )

print("The Walmart dataset exists.")
```

```python
# Check the file size

file_size_bytes = raw_data_path.stat().st_size

file_size_mb = file_size_bytes / (1024 ** 2)

print("Filename:", raw_data_path.name)
print(f"File size: {file_size_mb:.2f} MB")
```

```python
# Load only a preview first
data_preview = pd.read_csv(
    raw_data_path,
    nrows=5,
)

data_preview
```

```python
# Inspect the column names

data_preview.columns.tolist()
```

```python
# Validata the expected columns

expected_columns = [
    "Store",
    "Date",
    "Weekly_Sales",
    "Holiday_Flag",
    "Temperature",
    "Fuel_Price",
    "CPI",
    "Unemployment",
]

actual_columns = data_preview.columns.tolist()

if actual_columns != expected_columns:
    raise ValueError(
        "The dataset columns do not match the expected structure.\n"
        f"Expected: {expected_columns}\n"
        f"Found: {actual_columns}"
    )

print("Dataset columns validated successfully.")
```

```python
# Load the complete dataset

df = pd.read_csv(raw_data_path)

print("Dataset columns validated successfully.")
```
