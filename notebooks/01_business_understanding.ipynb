# Project Overview

Retail businesses depend heavily on accurate sales forecasting to make informed business decisions.

This project aims to build a machine learning model capable of predicting Walmart's weekly sales using historical sales data and relevant features such as holidays, store information, fuel prices, temperature, umemployemnt rate, and consumer price index(CPI).

The final model will be deployed as a Streamlit web application and hosted on Microsoft Azure App Service, allowing users to make sales predictions through a user-friendly interface.

# Business Problem

Walmart operates hundreds of retail stores and must accurately forecast weekly sales to support business planning.

Inaccurate sales predictions can lead to:

- Overstocking products
- Stock shortages
- Increased  storage costs
- Lost sales opportunities
- Poor inventory planning

The goal of this project is to develop a machine learning model that predicts weekly sales accurately using historical sales and economic indicators.

# machine Learning Problem

This project is a `Supervised Machine Learning Regression` problem.

### Why Regression?

The target variables is `Weekly Sales`, which is a continous numerical value.

Examples of predicted values include:
- 12,350.55
- 95,110.20
- 231,980.75

Since the model predicts continous numbers rather than categories, this is a regression task.


# Dataset Description

The dataset used in this project contains historical Walmart weekly sales records collected from multiple stores.

The dataset includes information about store characteristics, economic indicators, holidays, and environmental conditions that may influence weekly sales.

## Target Variable

The target variable is:

- **Weekly_Sales**

This is the value that the machine learning model will predict.

## Input Features

The dataset contains the following features:

| Feature | Description |
|----------|-------------|
| Store | Unique store identifier |
| Date | Weekly sales date |
| Holiday_Flag | Indicates whether the week includes a holiday (1 = Yes, 0 = No) |
| Temperature | Average weekly temperature |
| Fuel_Price | Weekly fuel price |
| CPI | Consumer Price Index |
| Unemployment | Unemployment rate |

These features will be explored to understand how they influence Walmart's weekly sales.

# Risks and Limitations

Although this project aims to build an accurate predictive model, several limitations should be considered.

## Risks

- Missing values may reduce model performance.
- Outliers may negatively affect some regression algorithms.
- Economic conditions can change over time.
- Consumer buying behaviour may change unexpectedly.
- Holiday effects may differ across regions and years.

## Limitations

- The model only learns from the available historical data.
- Future events not represented in the dataset cannot be predicted.
- Model performance depends on the quality of the dataset.
- Predictions should support business decisions rather than replace human judgement.
